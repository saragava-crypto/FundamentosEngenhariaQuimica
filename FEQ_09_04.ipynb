```python
import numpy as np

soma_limite= np.pi/4
n = 0
s=[]
erro = 10**-6


while erro > 10**-6:
  termo= np.sum(((-1)**n)/(2*n+1))
  s.append(termo)
  soma=np.sum(s)
  erro =(soma-soma_limite)**2



```

```python
import numpy as np

tol=1e-6
soma=0
n=0

while True:
  S=((-1)**n/(2*n+1))
  soma += S
  erro=abs(soma-np.pi/4)
  if erro**2<tol:
    break
  n+=1

print("Número de termos necessários:", n + 1)
```
